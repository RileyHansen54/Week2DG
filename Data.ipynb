# Introduction
This Jupyter Notebook is prepared for the exploratory data analysis (EDA) of the cab industry investment opportunities for XYZ firm. The datasets provided include transaction records, customer demographics, and city statistics for the period from 31/01/2016 to 31/12/2018. The goal is to derive actionable insights that will inform XYZ's investment decision.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
print(os.getcwd())

%matplotlib inline
cab_data = pd.read_csv('python/Cab_Data.csv')
customer_id = pd.read_csv('python/Customer_ID.csv')
transaction_id = pd.read_csv('python/Transaction_ID.csv')
city_data = pd.read_csv('python/City.csv')

display(cab_data.head())
display(customer_id.head())
display(transaction_id.head())
display(city_data.head())
print(cab_data.info())
print(customer_id.info())
print(transaction_id.info())
print(city_data.info())
print(cab_data.isnull().sum())
print(customer_id.isnull().sum())
print(transaction_id.isnull().sum())
print(city_data.isnull().sum())
print(cab_data.describe())
print(city_data.describe())
cab_data.dropna(inplace=True)
city_data.ffill(inplace=True)
cab_data.drop_duplicates(inplace=True)
cab_data['Date of Travel'] = pd.to_datetime(cab_data['Date of Travel'], unit='D', origin='1899-12-30')

plt.figure(figsize=(10, 6))
sns.histplot(cab_data['Price Charged'], kde=True)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(cab_data['Price Charged'], cab_data['KM Travelled'])
plt.title('Price Charged vs KM Travelled')
plt.xlabel('KM Travelled')
plt.ylabel('Price Charged')
plt.show()
```
